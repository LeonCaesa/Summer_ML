```python
def split(dataset, month_of_looking_back = 12):
    '''Ask user to enter a date, and make sure the date is included in the dataset.'''
    while dataset['DATE'].dtypes != 'object':
        dataset['DATE']=dataset['DATE'].dt.strftime('%Y-%m').astype('object')
    date_array = dataset.DATE.unique()
    date_list = date_array.tolist()
    print('List of all date:')
    print(date_list)
        
    date = input("Enter your date: ")
    while date not in date_list:
        print("Invalid date, please enter a valid date. ")
        date = input("Enter your date: ")
        
    
    '''Now we have the split time.'''
    time  = pd.to_datetime(date, format = '%Y-%m')
    while dataset['DATE'].dtypes != 'datetime64[ns]':
        dataset['DATE']=pd.to_datetime(dataset['DATE'], format = '%Y-%m')
    
    
    '''Set looking backward dataset/train dataset.'''    
    time1 = time - relativedelta(months=month_of_looking_back)
    time2 = time1.strftime('%Y-%m')
    
    '''Check if looking back period has gone out of range.'''
    while time2 not in date_list:
        print("Invalid date, please enter a valid date. ")
    
    train = dataset.loc[(dataset['DATE']>time2) & (dataset['DATE']<=time)]
    
    '''Set looking ahead dataset/test dataset.'''
    time3 = time + relativedelta(months=1)
    time4 = time3.strftime('%Y-%m')
    
    '''Check if looking ahead period has gone out of range.'''
    while time4 not in date_list:
        print("Invalid date, please enter a valid date. ")
    test = dataset.loc[dataset['DATE']==time4]
    
    '''Drop gvkeys, date and all other irrelevant columns.'''
    x_train = train[train.columns[5:-1]]
    y_train = train[train['DATE']==time]['RET']
    
    x_test = test[test.columns[5:-1]]
    y_test = test['RET']
       
    print('shape of y_train: (%s,)'%y_train.shape)
    print('shape of x_train: (%s, %s,)'%x_train.shape)

    print('shape of y_test: (%s,)'%y_test.shape)
    print('shape of x_test: (%s, %s,)'%x_test.shape)
    
    
    return x_train, y_train, x_test, y_test
        
    
    
    
```

```python
# Output all possible split date and ask user to pick one.
def split_date(dataset):
    while dataset['DATE'].dtypes != 'object':
        dataset['DATE']=dataset['DATE'].dt.strftime('%Y-%m-%d').astype('object')
    date_array = dataset.DATE.unique()
    date_list = date_array.tolist()
    print('List of all date:')
    print(date_list)
        
    date = input("Enter your date: ")
    while date not in date_list:
        print("Invalid date, please enter a valid date. ")
        date = input("Enter your date: ") 
    return date
```

```python
# Input: a cleaned dataset, with 'DATE' indicating date of variable, dataset is sorted in timely order.
# y variable split date.
# train date ends.
# Fixed training period.
# Output: train_et, test_set.
def split_data(dataset,  y_split, month = 3):
    

    time  = pd.to_datetime(y_split, format = '%Y-%m-%d')
    while dataset['DATE'].dtypes != 'datetime64[ns]':
        dataset['DATE']=pd.to_datetime(dataset['DATE'], format = '%Y-%m-%d')
    
    # Shift month extra months ahead to include training data.
    time1 = time - relativedelta(months=month)
    time2 = time1.strftime('%Y-%m-%d')
    
    # train dataset in df1:
    df1 = dataset.loc[dataset['DATE']<=y_split]
    # test dataset in df2:
    df2 = dataset.loc[dataset['DATE']>=time2]
    
    return df1, df2        
```

```python
# Input: train dataset, month of looking ahead.
# Output: An organized trainset => x_train, y_train. x_train 3-d, only proper for LSTM model training.


# Parameter of month should align with the parameter in the last function.
# All companies trained in one model; dataset is already sorted in the timely order.
# x_variables: 3-month data related to that one particular company.
# y_variable: one month ahead return.

# Some variables repeat every several periods - make it hard to train models.

def train_set(train, month=3):
    # y trainset.
    y_variable = []
    # x trainset.
    x_variable = []
    
    # First identify companies, and append x,y variables respectively in a timely order.
    train_co_list = train.gvkey.unique().tolist()
    for co in train_co_list:
        co_df = train[train['gvkey'] == co]
        count_row = co_df.shape[0]
        x = co_df[co_df.columns[5:-1]]
        
        if count_row > month:
            for i in range(month, count_row):
                y_variable.append(co_df.iloc[i,-1])
                # x_var = x.iloc[i-month:i, ].values
                x_variable.append(x.iloc[i-month:i, ].values)
    print(train.shape[0])
    
    x_train = np.asarray(x_variable)
    y_train = np.asarray(y_variable)
    # y_train = np.asarray(np.array(y_variable))

    
    print('shape of y_train: (%s,)'%y_train.shape)
    print('shape of x_train: (%s, %s,%s,)'%x_train.shape)
    
    return x_train, y_train   
```

```python
# scaler transformer required later.
def test_set(test, month=3):
    y_variable = []
    # x testset.
    x_variable = []
    
    # First identify companies, and append x,y variables respectively in a timely order.
    test_co_list = test.gvkey.unique().tolist()
    for co in test_co_list:
        co_df = test[test['gvkey'] == co]
        count_row = co_df.shape[0]
        x = co_df[co_df.columns[5:-1]]
        
        if count_row > month:
            for i in range(month, count_row):
                y_ = [co, co_df.DATE.values[i], co_df.iloc[i,-1]]
                y_variable.append(y_)
                # x_var = x.iloc[i-month:i, ].values
                x_variable.append(x.iloc[i-month:i, ].values)
    print(test.shape[0])
    
    '''Scaler function here.'''
    x_test = np.asarray(x_variable)
    y_test = np.asarray(y_variable)
    # y_train = np.asarray(np.array(y_variable))

    
    # print('shape of y_train: (%s,)'%y_test.shape)
    print('shape of x_test: (%s, %s,%s,)'%x_test.shape)
    
    return x_test, y_test    
```

```python
# LSTM Model in particular.
# Can be substituted by any other models.
def LSTM_model(x_train, y_train):
    import tensorflow.keras.optimizers as optimizers
    from tensorflow.keras.layers import LSTM, Dense, Activation, Dropout
    from tensorflow.keras.models import Sequential
    dropout_rate = 0.45

    model = Sequential()

    model.add(LSTM(units=160, return_sequences=True))
    model.add(Dropout(rate=dropout_rate))
    model.add(LSTM(units=80, return_sequences=False))
    model.add(Dropout(rate=dropout_rate))
    model.add(Dense(units=80))
    model.add(Activation("relu"))
    model.add(Dropout(rate=dropout_rate))
    model.add(Dense(units=1))
    model.add(Activation("linear"))
    rms=optimizers.RMSprop(lr=2e-3, rho=0.9, epsilon=1e-06)

    start = time.time()
    model.compile(loss="mse", optimizer=rms)
    print ('compilation time : ', time.time() - start)
    fitted = model.fit(x=x_train, y=(y_train), epochs =50, batch_size=3000)
    return fitted
```
